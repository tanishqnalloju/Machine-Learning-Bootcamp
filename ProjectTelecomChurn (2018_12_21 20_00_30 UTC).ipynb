```python
import numpy as np
import pandas as pd
data = pd.read_csv('Salary_Data.csv')
#correaltionship strength
np.corrcoef(data['YearsExperience'],data['Salary'])
```

```python
#model building
#x = input variable that contain work experience
x = data[['YearsExperience']]
y = data[['Salary']]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression().fit(x, y)
```

```python
print(lr)
```

```python
#score tells the accuracy of a model
lr.score(x, y)
```

```python
lr.predict(2.3)
```

```python
lr.predict(7.1)
```

```python
#intercept  B0
lr.intercept_
```

```python
#slope B1
lr.coef_
```

```python
#LR formula
#y = B0+B1*x
data['predicted_salary'] = lr.predict(x)
```
